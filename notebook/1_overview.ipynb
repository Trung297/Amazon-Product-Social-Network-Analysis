# amazon-meta

| Thuộc tính           | Kiểu dữ liệu | Mô tả                                                                                                          |
| -------------------- | ------------ | -------------------------------------------------------------------------------------------------------------- |
| `Id`                 | Integer      | Mã định danh nội bộ của sản phẩm trong tập dữ liệu.                                                            |
| `ASIN`               | String       | Amazon Standard Identification Number (ASIN), mã định danh duy nhất của sản phẩm trên Amazon.                  |
| `title`              | String       | Tên hoặc tiêu đề của sản phẩm.                                                                                 |
| `group`              | String       | Nhóm sản phẩm (ví dụ: Book, Music, DVD, Video, Software, Toy, Video Games, ...).                               |
| `salesrank`          | Integer      | Thứ hạng bán hàng của sản phẩm trên Amazon. Giá trị càng nhỏ cho thấy sản phẩm bán chạy hơn.                   |
| `similar`            | List[String] | Danh sách ASIN của các sản phẩm tương tự được Amazon đề xuất. Số đầu tiên biểu thị số lượng sản phẩm tương tự. |
| `categories`         | List[String] | Danh sách các danh mục phân loại của sản phẩm dưới dạng cây phân cấp (hierarchical categories).                |
| `reviews.total`      | Integer      | Tổng số lượng đánh giá (reviews) của sản phẩm.                                                                 |
| `reviews.downloaded` | Integer      | Số lượng đánh giá chi tiết được tải về trong tập dữ liệu.                                                      |
| `reviews.avg_rating` | Float        | Điểm đánh giá trung bình của sản phẩm (thang điểm 1–5).                                                        |
| `review.date`        | Date         | Ngày đăng đánh giá của người dùng.                                                                             |
| `review.customer`    | String       | Mã định danh của khách hàng thực hiện đánh giá.                                                                |
| `review.rating`      | Integer      | Điểm đánh giá của khách hàng (1–5 sao).                                                                        |
| `review.votes`       | Integer      | Tổng số lượt bình chọn cho đánh giá.                                                                           |
| `review.helpful`     | Integer      | Số lượt bình chọn đánh giá là hữu ích.                                                                         |
